# Preparing Data

```python
import json
import pandas as pd
import numpy as np
```

```python
# read snli

with open('data/chaosNLI/data/chaosNLI_v1.0/chaosNLI_snli.jsonl', 'r') as f:
    data = f.read()
data_dict = []
c = 0
for d in data.split('\n')[:-1]:
    data_dict.append(json.loads(d))
df_snli_full = pd.DataFrame(data_dict)
```

```python
# read mnli

with open('data/chaosNLI/data/chaosNLI_v1.0/chaosNLI_mnli_m.jsonl', 'r') as f:
    data = f.read()
data_dict = []
c = 0
for d in data.split('\n')[:-1]:
    data_dict.append(json.loads(d))

df_mnli_full = pd.DataFrame(data_dict)
```

```python
# combine and randomise

df_chaos_full = pd.concat([df_snli_full, df_mnli_full])
df_chaos_full = df_chaos_full.sample(frac=1, random_state=100).reset_index(drop=True)

# adding p h columns
df_chaos_full['premise'] = [x['premise'] for x in df_chaos_full['example']]
df_chaos_full['hypothesis'] = [x['hypothesis'] for x in df_chaos_full['example']]
df_chaos_full.head()
```

```python
df_chaos_train = df_chaos_full[:-1000]
df_temp = df_chaos_full[-1000:]
df_chaos_val = df_temp[:-500]
df_chaos_test = df_temp[-500:]
assert len(df_chaos_train)+len(df_chaos_val)+len(df_chaos_test) == len(df_chaos_full)
```

```python
# save to file

df_chaos_train.to_csv('data/df_chaos_train.csv', index=False)
df_chaos_val.to_csv('data/df_chaos_val.csv', index=False)
df_chaos_test.to_csv('data/df_chaos_test.csv', index=False)
```

```python
def count_labels(labels):
    e = 0
    n = 0
    c = 0
    for l in labels:
        if l == 'entailment':
            e += 1
        elif l == 'neutral':
            n += 1
        elif l == 'contradiction':
            c += 1
        elif l == '':
            continue
        else:
            raise Exception('Label error')
    return [e, n, c]  

def label_counts_to_dist(counts):
    a = np.array(counts)
    dist = a / np.sum(a)
    return dist.tolist()
```
